```
import random
import numpy as np

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
def calcula_pi(n, print_results=True):
  a, b = -1., 1.
  xs = np.random.uniform(a, b, n) 
  ys = np.random.uniform(a, b, n) 

  en_circulo = np.where((np.sqrt(xs**2 + ys**2)<=1), 1, 0)

  puntos_circulo = (en_circulo==1).sum()
  puntos_cuadrado = n

  pi = 4* puntos_circulo/ puntos_cuadrado
  error = abs(np.pi-pi)

  if print_results==True:
    plt.figure(figsize=(6,6))
    sns.scatterplot( x=xs, y=ys, s = 4,hue=en_circulo )
    plt.show()

    print('pi =', pi)
    print('error =', error)

  return pi
```

```
calcula_pi(1000000)
```

```
ns = range(10, 100000, 100)
pi_est = []
for n in ns:
  pi = calcula_pi(n, print_results=False)
  pi_est.append(pi)

#Gráfica estimación de  pi vs pi
plt.figure(figsize=(16, 6))
plt.plot(ns, pi_est, color='g')
plt.axhline(y=np.pi, color='black', ls='--')
plt.xlim(0, ns[-1])
plt.ylim(2.2, 4)
plt.xlabel('n')
plt.ylabel('Estimación de pi')
plt.title('Estimación de pi según cantidad de puntos')
plt.show()
```

# ¿Podemos usar una distribución gaussiana o una distribución de laplace para la generación de nuestros puntos aleatorios?

```
n = 1000000
a, b = -1., 1.
xs= np.random.uniform(a, b, n) 
ys= np.random.uniform(a, b, n) 

sns.jointplot(xs, ys, kind = 'scatter', color = 'darkcyan', s = 2 )
plt.show()
```

```
mu, sigma = 0., 1/3
xs = np.random.normal(mu, sigma, n)
ys = np.random.normal(mu, sigma, n)

sns.jointplot(xs, ys, kind = 'scatter', color = 'darkcyan', s = 2 )
plt.show()
```

```
mu, scale = 0., 1/7 
xs = np.random.laplace(mu, scale, n)
ys = np.random.laplace(mu, scale, n)

sns.jointplot(xs, ys, kind = 'scatter', color = 'darkcyan', s = 2 )
plt.show()
```
